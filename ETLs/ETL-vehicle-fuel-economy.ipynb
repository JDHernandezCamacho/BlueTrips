```python
# Filtro de automoviles donde el combustible sea electricidad
df_filter_electricos = df.loc[df['fuelType1'].str.contains('Electricity', case=False, na=False)]
# Eliminando columnas con valores nulos
df_filter_electricos = df_filter_electricos.dropna(axis=1, how='all')
df_filter_electricos
```

```python
# Filtro de automoviles donde el combustible no sea electricidad
df_filter_no_electricos = df.loc[~df['fuelType1'].str.contains('Electricity', case=False, na=False)]
# Eliminando columnas con valores nulos
df_filter_no_electricos = df_filter_electricos.dropna(axis=1, how='all')
df_filter_no_electricos
```
